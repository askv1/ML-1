### Pipeline Version (2 points)

```python
# 2 points
from sklearn.base import BaseEstimator, TransformerMixin
class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, copy=True, with_mean=True, with_std=True):
        # your code here
        self.copy = copy
        self.with_mean = with_mean
        self.with_std = with_std

    def fit(self, X, y=None):
        # your code here
        self.std = X.std(axis=0, ddof=0)
        self.mean = X.mean(axis=0)     
        return self

    def transform(self, X, y=None, copy=None):
        # your code here
        if self.copy:
            return ((X - self.mean) / self.std).values
        
        if self.with_mean:
            X -= self.mean
        if self.with_std:
            X /= self.std
        return X
        
```

```python
assert np.allclose(CustomScaler().fit_transform(wine_data), StandardScaler().fit_transform(wine_data))
```

## Exercise 4 - manual PCA (5 points)
The task is to solve PCA as an optimization problem, without explicitly doing eigen value decomposition.
In the most general setting PCA is minimization of reconstruction error of a projection of given rank $q$

$$\min_{\mu, \lambda_1,\ldots, \lambda_n, \mathbf{V}_q} \sum_{i=1}^n ||x_i - \mu - \mathbf{V}_q \lambda_i||^2$$

With a number of steps that can be found here https://stats.stackexchange.com/a/10260 this task transforms to
 $$\max_{u_i} \sum_{i=1}^q u_i^T \mathbf{S} u_i$$
 where $\mathbf{S}$ is the sample covariance matrix (after standartization) and $u_1, \ldots, u_q$ are the $q$ are orthonormal columns in $\mathbf{V}_q$.
 Let us solve this optimization problem with `scipy.optimize` library.
 
 Additional 2 point are given for visualization of the results.

### PCA (3 points)

```python
wine_data, wine_labels = wine_sklearn['data'], wine_sklearn['target']
```

Find a covariance matrix of standartized data and assing it to S.

```python
## your code

wine_data = StandardScaler().fit_transform(wine_data)

#S = np.cov(wine_data.T, bias=False)

S = (wine_data.T @ wine_data) / (wine_data.shape[0] - 1)
```

If your code is correct, the following assert should be Ok.

```python
assert np.allclose(np.linalg.norm(S), 5.787241159764733)
```

```python
from scipy.optimize import minimize

def objective(x):
    # your code: write objective of the problem (don't forget that scipy does min while we need max)
    return -1 * (x @ S @ x)

def norm_constraint(x):
    # your code: constaraint norm of x to be 1, function should return 0 if constraint holds
    return x @ x - 1
    
con1 = {'type': 'eq', 'fun': norm_constraint}

x0 = np.full(len(S), 0.5) - np.random.random(len(S))
#x0 = x0 / (x0 @ x0) ** 0.5

sol = minimize(objective, 
               x0.T, 
               constraints = [con1]
              )
x0 = sol.x
```

Hurray! We have first vector! Let's do another one.

```python
def orthogonality_constraint(x):
    # your code: x should be orthogonal to x0, function should return 0 if constraint holds
    return x @ x0

con2 = {'type': 'eq', 'fun': orthogonality_constraint}

x1 = np.full(len(S), 0.5) - np.random.random(len(S))

sol = minimize(objective, 
               x1, 
               constraints = [con1, con2]
              )

x1 = sol.x
```

If your solution is correct, the following asserts should be Ok.

```python
assert np.allclose(x0@S@x0, 4.732436977583595)
assert np.allclose(x1@S@x1, 2.5110809296451233)
```

### Visualization (2 points)

Visualize the points after applying custom dimension reduction with 2 components.

```python
pc1 = wine_data @ x0
pc2 = wine_data @ x1

plot_scatter(pc1, pc2, clusters=wine_labels, auto_scaled=False)
```

## Exercise 5 - Boruta (3 points)

Let us classify handwritten digits 0, 1 and 2. 
To make task not so easy the images are binarized (no shadows of gray present) as it happens with xerocopied documents.

Let us also find out to which parts of an image there's no need to look in order to clasify three digits of interest.

```python
X, y = load_digits(n_class=3, return_X_y=True, as_frame=True)
X = (X>10).astype(int)
```

```python
f, ax = plt.subplots(1,3,figsize=(10,4))
for i in range(3):
    ax[i].imshow(X.iloc[i].values.reshape(8,8))
    ax[i].set_title(f"This is digit {y[i]}.")
plt.suptitle("First three images.")
plt.show()
```

Split data into train and test, let test size be 30% of the dataset and fix random state to 42:

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
assert y_val.shape[0] == 162
assert y_val.sum() == 169
```

Fit a RandomForestClassifier with max_depth=13 and evaluate it's performance:

```python
clf = RandomForestClassifier(max_depth=13, random_state=42)

# your code here
clf.fit(X_train, y_train)

acc = np.mean(clf.predict(X_val) == y_val)
print(acc)
```

```python
assert acc > 0.98
```

```python
from boruta import BorutaPy

feat_selector = BorutaPy(RandomForestClassifier(max_depth=13), 
                         n_estimators='auto', 
                         verbose=0, 
                         max_iter=100,
                         random_state=42)

# your code here: do run the boruta

feat_selector.fit(X_train.values, y_train)
```
